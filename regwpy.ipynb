```python
# Import the libraries
import pandas as pd
import numpy as np
```

```python
# Read file, show header
path='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'
df = pd.read_csv(path)
df.head
```

```python
# Import libraries with visualizations
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Show data types
print(df.dtypes)
```

```python
# Show correlations from data
df.corr()
```

```python
# Create dataframe with specific categories
df[['bore', 'stroke', 'compression-ratio', 'horsepower']].corr()
```

```python
# Plot specific categories on graph (Consumers are willing to pay about 50% more for double the horsepower, but they are outliers)
sns.regplot(x="engine-size", y="price", data=df)
plt.ylim(0,)
```

```python
# Show correlation on engine size, price (Here the correlation is strong, really close to 1) 
df[["engine-size", "price"]].corr()
```

```python
# Plot highway mpg and price on chart (General distribution is around $8k - $20k for cars with higher mpg, drivers that pay more for their cars generally have lower mpg or if they drive like me they get less mpg :) 
sns.regplot(x="highway-mpg", y="price", data=df)
```

```python
# Show correlation between stroke and price (Here the correlation is weak, closer to 0) 
df[["stroke", "price"]].corr()
```

```python
# Example on how there is no correlation, no fitted values, no consensus
sns.regplot(x="stroke", y="price", data=df)
```
